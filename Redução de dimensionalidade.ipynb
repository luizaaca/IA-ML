```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image

def load_image(image_path):
    """Load an image from a file."""
    with open(image_path, "rb") as file:
        img = Image.open(file)
        img = img.convert("RGB")  # Ensure the image is in RGB format
    return img

def convert_to_grayscale(img):
    """Convert an RGB image to grayscale."""
    width, height = img.size
    grayscale_img = []

    for y in range(height):
        row = []
        for x in range(width):
            r, g, b = img.getpixel((x, y))
            # Use standard grayscale formula: 0.299*R + 0.587*G + 0.114*B
            gray = int(0.299 * r + 0.587 * g + 0.114 * b)
            row.append(gray)
        grayscale_img.append(row)

    return grayscale_img

def apply_threshold(grayscale_img, threshold=127):
    """Convert a grayscale image to binary using a fixed threshold."""
    binary_img = []
    for row in grayscale_img:
        binary_row = [255 if pixel > threshold else 0 for pixel in row]
        binary_img.append(binary_row)
    return binary_img


def process_image(image_path):
  # Load the image
  img = load_image(image_path)

  # Convert to grayscale
  grayscale_img = convert_to_grayscale(img)

  # Convert to binary
  binary_img = apply_threshold(grayscale_img, threshold=107)

  # Display the images
  plt.figure(figsize=(12, 4))

  plt.subplot(1, 3, 1)
  plt.imshow(img)
  plt.title("Original Image")
  plt.axis('off')

  plt.subplot(1, 3, 2)
  plt.imshow(grayscale_img, cmap='gray')
  plt.title("Grayscale Image")
  plt.axis('off')

  plt.subplot(1, 3, 3)
  plt.imshow(binary_img, cmap='gray')
  plt.title("Binary Image")
  plt.axis('off')

  plt.show()

# Example usage
process_image("image1.jpg")
```
